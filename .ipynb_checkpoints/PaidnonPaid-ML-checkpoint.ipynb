```python
dt.drop(['Spouse Employment  status - Employed/unemployed','Duration of credit'], axis = 1, inplace = True)
```

```python
dt.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
dt['Educational level - literate/Illiterate'] = le.fit_transform(dt['Educational level - literate/Illiterate'])
```

```python
le2 = LabelEncoder()
dt['Nature of livelihood - Farm/Non Farm'] = le2.fit_transform(dt['Nature of livelihood - Farm/Non Farm'])
```

```python
min_val = dt['Amount of credit'].min()
max_val = dt['Amount of credit'].max()
maxmin = max_val - min_val
for i in range(len(dt['Amount of credit'])):
    dt['Amount of credit'][i] = (dt['Amount of credit'][i] - min_val)/maxmin
```

```python
dt.head()
```

```python
min_val1 = dt['Rate of Interest'].min()
max_val1 = dt['Rate of Interest'].max()
maxmin1 = max_val1 - min_val1
for i in range(len(dt['Rate of Interest'])):
    dt['Rate of Interest'][i] = (dt['Rate of Interest'][i] - min_val1)/maxmin1
```

```python
dt.head()
```

```python
dt_2 = dt.drop(['Sector of livelihood', 'Purpose of credit', 'Paid /Non Paid (Default)'], axis = 1)
y = dt['Paid /Non Paid (Default)']
```

```python
le3 = LabelEncoder()
dt['Paid /Non Paid (Default)'] = le3.fit_transform(dt['Paid /Non Paid (Default)'])
```

```python
corr_matrix = dt.corr()
corr_matrix['Paid /Non Paid (Default)'].sort_values(ascending=False)
```

```python
dt_2.head()
```

```python
x = dt.drop(['age_grp', 'No of dependent family members', 'Paid /Non Paid (Default)'], axis = 1)
```

```python
x.head()
```

```python
ordinal_label = {k : i for i, k in enumerate(x['Sector of livelihood'].unique(),0)}
x['Sector of livelihood'] = x['Sector of livelihood'].map(ordinal_label)
```

```python
ordinal_label2 = {l : j for j, l in enumerate(x['Purpose of credit'].unique(),0)}
x['Purpose of credit'] = x['Purpose of credit'].map(ordinal_label2)
```

```python
x.head()
```

```python
x['Purpose of credit'].isna().sum()
```

```python
from sklearn.feature_selection import chi2
fpvalue = chi2(x,y)
fpvalue
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
le4 = LabelEncoder()
y = le4.fit_transform(y)
```

```python
y
```

```python
y = y.reshape(len(y), 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
c1 = LogisticRegression(random_state=0)
c1.fit(x_train, y_train)
```

```python
y_pred = c1.predict(x_test)
```

```python
y_pred_train = c1.predict(x_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
cm2 = confusion_matrix(y_train, y_pred_train)
acc1 = accuracy_score(y_test, y_pred)
acc2 = accuracy_score(y_train, y_pred_train)
```

```python
print(acc1, acc2)
```

```python
from sklearn.svm import SVC
c2 = SVC(kernel = 'rbf', random_state = 0)
c2.fit(x_train, y_train)
```

```python
y_pred = c2.predict(x_test)
y_pred_train = c1.predict(x_train)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm2 = confusion_matrix(y_train, y_pred_train)
acc1 = accuracy_score(y_test, y_pred)
acc2 = accuracy_score(y_train, y_pred_train)
print(acc1, acc2)
```

```python
from sklearn.tree import DecisionTreeClassifier
c3 = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
c3.fit(x_train, y_train)
```

```python
y_pred = c3.predict(x_test)
y_pred_train = c3.predict(x_train)
cm = confusion_matrix(y_test, y_pred)
cm2 = confusion_matrix(y_train, y_pred_train)
acc1 = accuracy_score(y_test, y_pred)
acc2 = accuracy_score(y_train, y_pred_train)
print(acc1, acc2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
c4 = KNeighborsClassifier(n_neighbors=3, metric='minkowski', p=2)
c4.fit(x_train, y_train)
```

```python
y_pred = c4.predict(x_test)
y_pred_train = c4.predict(x_train)
cm = confusion_matrix(y_test, y_pred)
cm2 = confusion_matrix(y_train, y_pred_train)
acc1 = accuracy_score(y_test, y_pred)
acc2 = accuracy_score(y_train, y_pred_train)
print(acc1, acc2)
```

```python
from sklearn.naive_bayes import GaussianNB
c5 = GaussianNB()
c5.fit(x_train, y_train)
```

```python
y_pred = c5.predict(x_test)
y_pred_train = c5.predict(x_train)
cm = confusion_matrix(y_test, y_pred)
cm2 = confusion_matrix(y_train, y_pred_train)
acc1 = accuracy_score(y_test, y_pred)
acc2 = accuracy_score(y_train, y_pred_train)
print(acc1, acc2)
```

```python
from sklearn.ensemble import RandomForestClassifier
c6 = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
c6.fit(x_train, y_train)
```

```python
y_pred = c6.predict(x_test)
y_pred_train = c6.predict(x_train)
cm = confusion_matrix(y_test, y_pred)
cm2 = confusion_matrix(y_train, y_pred_train)
acc1 = accuracy_score(y_test, y_pred)
acc2 = accuracy_score(y_train, y_pred_train)
print(acc1, acc2)
```
